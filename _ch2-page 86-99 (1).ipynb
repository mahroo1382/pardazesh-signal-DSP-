## صفحه 90 - سیستم های خطی تغییر ناپذیر با زمان

تنها باید ببینیم با  میل به صفر ، یعنی ، چه رخ میدهد برای (1) بیان شده در معادله (۲-۲۶) ،  تقریب بهتر و بهتری از میشود و در این دو یکی میشوند. پس پاسخ یعنی   معادله (۲-۲۹) نیز باید ، چنانچه در شکل ۲-۱۵) (و) نشان داده شده است به ، پاسخ سیستم به  ورودی واقعی x(t) همگرا شود. به علاوه همانطور که گفتیم به ازای ۱۵ به حد کافی کوچک ، طول پالس  دیگر چندان مهم نیست از این نظر که پاسخ سیستم به این پالس ، اساساً با پاسخ سیستم به  ضربه واحدی در همین زمان برابرست. یعنی چون پالس به ازای با یک ضربه واحد  جابجا شده متناظر ست ، پاسخ به این پالس در حد به پاسخ ضربه می گراید. پس اگر پاسخ سیستم به ضربه باشد، آنگاه  (2-30)  

$$  y(t) = \lim_{\Delta \to 0} \sum_{k=-\infty}^{\infty} x(k\Delta) h_{k\Delta}(t) \Delta $$

به ازای چنانچه شکل ۲-۱۶ نشان میدهد جمع سمت راست به یک انتگرال تبدیل می شود . در  شکل ۲-۱۶ مستطیل سایه دار یکی از جملات جمع سمت راست معادله را نشان می دهد ، و به ازای  ۰ - ۵ این جمع به مساحت زیر به عنوان تابعی از ۲ ، میل میکند . بنابراین  (2-31)   

$$ y(t) = \int_{-\infty}^{\infty} x(\tau) h_{t}(\tau) d\tau $$

معادلة (۲-۳۱) تعبیری همانند تعبیر معادله (۲-۲۹) دارد. چنانچه در بخش ۲-۲-۱ نشان داده شد، هر  ورودی میتوان به صورت زیر بیان کرد  

$$ x(t) = \int_{-\infty}^{\infty} x(\tau) \delta (t - \tau) d\tau $$

یعنی میتوانیم مجموع وزندار ضربه های جابجا شده ای تصور کنیم که هر ضربه برابر (۲) است . معادله (2-31) با این تعبیر مجموع پاسخهای سیستم به این ورودیهاست، و طبق اصل  خطی بودن وزن پاسخ به ضربة جابجا شده نیز باید باشد 
معادله (۲-۳۱) شکل کلی پاسخ یک سیستم خطی پیوسته در زمان را نشان می دهد. اگر سیستم خطی  تغییر ناپذیر با زمان هم باشد ، داریم یعنی پاسخ سیستم خطی تغییر ناپذیر با زمان به ضربه واحد جابجا  شده که به اندازه 1 ثانیه نسبت به مبدا جابجا شده است پاسخ سیستم به ضربه واحد است  

```python
## شکل 2-16

import numpy as np
import matplotlib.pyplot as plt

# تعریف بازه زمانی
t = np.linspace(0, 2, 400)

# مقادیر دلتا و نقاط نمونه‌برداری شده
delta = 0.2
k = 0.8  # مقدار k انتخاب شده برای نشان دادن ضرب سیگنال

# تعریف سیگنال x(t)
def x_t(t):
    return np.sin(2 * np.pi * t)  # سیگنال ورودی سینوسی

# تعریف تابع پاسخ ضربه‌ای h_kΔ(t)
def h_k(t, k):
    return np.exp(-5 * (t - k)**2) * np.sin(4 * np.pi * (t - k))

# محاسبه ضرب x(kΔ)h_kΔ(t)
x_k = x_t(k)  # مقدار نمونه‌برداری شده x در kΔ
y_t = x_k * h_k(t, k)  # ضرب سیگنال در پاسخ ضربه‌ای

# رسم شکل
fig, ax = plt.subplots(figsize=(6, 4))

# رسم سیگنال h_kΔ(t)
ax.plot(t, h_k(t, k), linestyle="dashed", color='black', alpha=0.6, label=r"$h_k(t)$")

# رسم ناحیه پر شده که نشان‌دهنده ضرب سیگنال ورودی و پاسخ است
ax.fill_between(t, 0, y_t, color='gray', alpha=0.4, label=r"$x(k\Delta) h_k(t) \Delta$")

# رسم محورهای اصلی
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# نمایش خطوط عمودی در kΔ و (k+1)Δ
ax.axvline(k, color='black', linestyle='dotted')
ax.axvline(k + delta, color='black', linestyle='dotted')

# برچسب محور x
ax.set_xticks([k, k + delta])
ax.set_xticklabels([r"$k\Delta$", r"$(k+1)\Delta$"])

# برچسب محورها
ax.set_xlabel(r"$t$")
ax.set_ylabel("Amplitude")
ax.legend()

plt.show()
```

که به همان اندازه جابجا شده باشد. باز هم برای سهولت زیر نویس ۸۰ را حذف کرده ، را به عنوان  پاسخ ضربه واحد تعریف میکنیم

در شکل ۲-۱۷ توابع به ازای یک مقدار منفی و یک مقدار مثبت ، رسم  شده اند . با توجه به این شکل می بینیم که به ازای حاصل ضرب صفر و در نتیجه هم صفرست. به ازای (5-44)

```python
## کد محور بالای صفحه 92 - شکل 2-17 

import numpy as np
import matplotlib.pyplot as plt

# ایجاد بردار زمان
tau = np.linspace(0, 5, 100)

# توابع موردنظر
h_tau = np.where(tau >= 0, 1, 0)  # h(τ)
x_tau = np.exp(-tau)  # x(τ)

# h(t - τ) برای t < 0
t_negative = np.where(tau < 0, 1, 0)

# h(t - τ) برای t > 0
t_positive = np.where(tau >= 0, 1, 0)

# رسم نمودارها
fig, axes = plt.subplots(4, 1, figsize=(6, 8), sharex=True)

# h(τ)
axes[0].plot(tau, h_tau, 'k', linewidth=2)
axes[0].set_ylabel(r"$h(\tau)$")
axes[0].axhline(0, color='black', linewidth=1)
axes[0].axvline(0, color='black', linewidth=1)
axes[0].set_ylim(-0.1, 1.1)

# x(τ)
axes[1].plot(tau, x_tau, 'k', linewidth=2)
axes[1].set_ylabel(r"$x(\tau)$")
axes[1].axhline(0, color='black', linewidth=1)
axes[1].axvline(0, color='black', linewidth=1)

# h(t - τ) برای t < 0
axes[2].plot(tau, t_negative, 'k', linewidth=2)
axes[2].set_ylabel(r"$h(t-\tau) \quad (t < 0)$")
axes[2].axhline(0, color='black', linewidth=1)
axes[2].axvline(0, color='black', linewidth=1)
axes[2].set_ylim(-0.1, 1.1)

# h(t - τ) برای t > 0
axes[3].plot(tau, t_positive, 'k', linewidth=2)
axes[3].set_ylabel(r"$h(t-\tau) \quad (t > 0)$")
axes[3].axhline(0, color='black', linewidth=1)
axes[3].axvline(0, color='black', linewidth=1)
axes[3].set_ylim(-0.1, 1.1)

# تنظیمات نهایی
axes[3].set_xlabel(r"$\tau$")
plt.tight_layout()
plt.show()
```

```python
## شکل محور 2-21 - y(t) = x(t) * h(t) سیگنال 

import numpy as np
import matplotlib.pyplot as plt

# مقدار T را مشخص می‌کنیم
T = 1  # می‌توان مقدار دیگری نیز انتخاب کرد

# محدوده زمانی
t = np.linspace(0, 3*T, 1000)  # 1000 نقطه بین 0 تا 3T

# مقداردهی اولیه y
y = np.zeros_like(t)  

# محاسبه مقادیر y بر اساس روابط داده شده
for i in range(len(t)):
    if 0 <= t[i] < T:
        y[i] = (1/T) * (t[i]**2)
    elif T <= t[i] < 2*T:
        y[i] = (T * t[i]) - ((1/T) * (t[i]**2))
    elif 2*T <= t[i] < 3*T:
        y[i] = -(1/T) * (t[i]*2) + (T * t[i]) + ((1/T) * (T*2))

# رسم نمودار
plt.figure(figsize=(6, 4))
plt.plot(t, y, 'k', linewidth=2, label=r'$y(t) = x(t) * h(t)$')
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("شکل ۲-۲۱: سیگنال $y(t) = x(t) * h(t)$ برای مثال ۷-۲")
plt.axhline(0, color='black', linewidth=0.5)  # خط محور x
plt.axvline(0, color='black', linewidth=0.5)  # خط محور y
plt.xticks([0, T, 2*T, 3*T], ['0', 'T', '2T', '3T'])  # تنظیم محور x
plt.yticks([])  # حذف مقیاس اعداد محور y
plt.grid(True, linestyle="--", alpha=0.6)  # نمایش خطوط راهنما
plt.legend()
plt.show()
```

با این تغییر متغیر نقش عوض می شود. طبق معادله (۲-۴۵) پاسخ یک سیستم خطی تغییر ناپذیر با زمان دارای  پاسخ ضربه به ورودی مشابه پاسخ یک سیستم خطی تغییر ناپذیر با زمان دارای پاسخ ضربه به ورودی  است. مثلاً برای محاسبه کانولوشن مثال ۲-۴ میتوانیم را وارونه و جابجا کنیم ، سیگنالهای را در هم ضرب کنیم و حاصلضرب را روی تمام مقادیر جمع ببندیم . 

به نحوی مشابه میتوانیم (۲-۴۴) را با تغییر متغیر اثبات کنیم، تعبیر این نتیجه نیز مانند تعبیر حالت  گسسته در زمان است : پاسخ یک سیستم خطی تغیر ناپذیر با زمان دارای پاسخ ضربه به ورودی مشابه پاسخ یک  سیستم دارای پاسخ ضربه  به ورودی است. پس میتوانیم کانولوشن مثال ۲-۷ را با وارونه و  جابجا کردن ضرب سیگنالهای  و انتگرالگیری از حاصلضرب در بیابیم. در بعضی موارد یکی از این دو صورت تجسم سیگنالها ساده ترست ، ولی همیشه هر دو صورت  جواب یکسانی به دست می دهند.

## صفحه 98 - سیستم های خطی تغییر ناپذیر با زمان
سیستمهای خطی تغسسر ناپذیر با زمان هستند و پاسخ ضربه شان داخل جعبه مشخص شده است. این نمایش ترسیمی راه مناسبی  برای مشخص کردن سیستمهای خطی تغییر ناپذیر با زمان در نمودارهای جعبه ای است و بار دیگر تأکید می کند که پاسخ ضربه  یک سیستم خطی تغییر ناپذیر با زمان به طور کامل رفتار آن سیستم را مشخص میکند .  

دو سیستم دارای پاسخ ضربه های ورودیهای یکسانی دارند، و خروجیهایشان باید با  هم جمع شوند. چون  
$$ y_1 (t) = x (t) * h_1 (t) $$
$$ y_2 (t) = x (t) * h_2 (t) $$
خروجی سیستم شکل ۲-۲۳ (الف) عبارت است از  (2-48)  
$$ y (t) = x (t) * h_1 (t) + x (t) * h_2 (t) $$
که با طرف راست معادله (۲-۴۷) متناظر ست. خروجی سیستم شکل ۲-۲۳ (ب) عبارت است از(2-49)   
$$ y (t) = x (t) * (h_1 (t) + h_2 (t)) $$
که با طرف چپ معادله (۲-۴۷) متناظر ست. با اعمال معادله (۲-۴۷) به معادله (۲-۴۹) و مقایسه نتیجه با  معادلة ( ۲-۴۸) در می یابیم که سیستمهای شکلهای ۲-۲۳ (الف) و (ب) هم ارزند .  

همتاهای گسسته در زمانشان را میگذاریم یعنی به جای را قرار میدهیم. پس به طور خلاصه ، به خاطر ترتیب  توزیع پذیر بودن کانولوشن روی جمع میتوان به جای ترکیب موازی دو سیستم خطی تغییر ناپذیر با زمان یک سیستم خطی تغییر ناپذری با زمان  قرار داد که پاسخ ضربه اش مجموع پاسخ ضربه های سیتمهای موازی شده باشد.  

همچنین نتیجه دیگر خواص جابجایی و توزیع پذیری این است که(2-50)    
$$ [x_1[n] + x_2[n]] * h[n] = x_1[n] * h[n] + x_2[n] * h[n] $$
(۵۱-۲)  
$$ [x_1 (t) + x_2 (t)] * h (t) = x_1 (t) * h (t) + x_2 (t) * h (t) $$
که میگوید پاسخ  یک سیستم خطی تغییر ناپذیر با زمان به مجموع دو ورودی باید با مجموع به پاسخ به تک تک ورودیها برابرباشد

```python
## کد محورشکل 2-24
import numpy as np
import matplotlib.pyplot as plt

# مقادیر محور n
n = np.arange(-3, 8)  # از -3 تا 7
y = np.piecewise(n, [n < 0, n >= 0], [0, lambda n: 3 + (n >= 1)])  # مقدار 3 برای n=0 و مقدار 4 از n=1 به بعد

# رسم نمودار سیگنال
plt.figure(figsize=(6,4))
plt.stem(n, y, basefmt="k")  # رسم سیگنال بدون use_line_collection
plt.axhline(0, color='black', linewidth=1)  # خط افقی
plt.axvline(0, color='black', linewidth=1)  # خط عمودی

# تنظیمات نمودار
plt.xticks(np.arange(-3, 9, 1))
plt.yticks([0, 1, 2, 3, 4])
plt.xlim(-3.5, 7.5)
plt.ylim(0, 5)
plt.xlabel("n")
plt.ylabel("y[n]")
plt.title("سیگنال y[n] مطابق شکل ۲-۲۴")
plt.grid(True, linestyle="--", alpha=0.6)

# نمایش نمودار
plt.show()
```
