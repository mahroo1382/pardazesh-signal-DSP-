که در آنها شکل ۲-۵ این سیگنالها را نشان میدهد. همچنین برای ساده شدن تجم و محاسبه کانولوشن سیگنالها ، در شکل ۲-۶ سیگنالهای $$ h[1-k] ,h[-1-k]  ,x[-k] $$ یعنی, سرانجام به ازای یک مثبت دلخواه و یک  منفی دلخواه نشان داده شده است. با توجه به این شکل می بینیم که در مقادير غير صفر و  روی هم نمی افتند.  


##  سیستمهای خطی تغییر ناپذیر با زمان صفحه 79
x[n] , h[n]  سیگنالها

$$ x[k]h[n-k]=a^k --  0≤k≤n $$
داریم n ≥ پس برای0  

$$ y[n]=\sum_{k=0}^{n} a ^ k  $$
به کمک نتیجه مسئله ۱-۵۴ می توان نوشت  
 
$$ y[n]=\sum_{k=0}^{n}\ a^k . 1-(a^n+1)/1-a $$

```python
## کد محور پایین صفحه 79 
import matplotlib.pyplot as plt
import numpy as np

# محدوده محور k
k = np.arange(-5, 6)  # از -5 تا 5

# تعریف سیگنال‌ها
h_k = np.where(k >= 0, 1, 0)  # h[k] = u[k] (تابع پله واحد)
h_minus_k = np.flip(h_k)  # h[-k]
h_minus_1_minus_k = np.roll(h_minus_k, -1)  # h[-1 - k]

# ایجاد نمودارها
plt.figure(figsize=(10, 8))

# نمودار اول: h[-k]
plt.subplot(3, 1, 1)
plt.stem(k, h_minus_k, basefmt=" ", use_line_collection=True)
plt.title('سیگنال h[-k]')
plt.xlabel('k')
plt.ylabel('h[-k]')
plt.grid(True)
plt.xticks(k)  # نمایش همه مقادیر k روی محور

# نمودار دوم: h[-1 - k]
plt.subplot(3, 1, 2)
plt.stem(k, h_minus_1_minus_k, basefmt=" ", use_line_collection=True)
plt.title('سیگنال h[-1 - k]')
plt.xlabel('k')
plt.ylabel('h[-1 - k]')
plt.grid(True)
plt.xticks(k)  # نمایش همه مقادیر k روی محور

# نمودار سوم: h[k] (برای مقایسه)
plt.subplot(3, 1, 3)
plt.stem(k, h_k, basefmt=" ", use_line_collection=True)
plt.title('سیگنال h[k]')
plt.xlabel('k')
plt.ylabel('h[k]')
plt.grid(True)
plt.xticks(k)  # نمایش همه مقادیر k روی محور

plt.tight_layout()
plt.show()
```
