## chapter5- Q1

## سؤال ۱: تبدیل فوریه گسسته در زمان
* تابع زیر داده شده است: $$ x[n] = δ[n] + δ[n-1] $$ تبدیل فوریه گسسته در زمان آن را بیابید

پاسخ:
با استفاده از تعریف DTFT: $$ X(e^{jω}) = Σ x[n] e^{-jωn} $$
داریم:
$$ X(e^{jω}) = 1 + e^{-jω} $$


## سؤال ۲: خواص تبدیل فوریه گسسته در زمان

اگر  $$ x[n] $$ تبدیل فوریه گسسته در زمان  $$ X(e^{jω}) $$ داشته باشد، تبدیل فوریه سیگنال  $$ y[n] = x[n-n0] $$ را بیابید.

* پاسخ:
با استفاده از خاصیت جابجایی در DTFT:
$$ Y(e^{jω}) = X(e^{jω}) e^{-jωn0} $$


## سؤال ۳: همبستگی در حوزه DTFT

اگر دو سیگنال  $$ x[n] $$  و $$ h[n] $$ داده شده باشند، تبدیل فوریه گسسته در زمان کانولوشن آن‌ها را مشخص کنید.

پاسخ:
طبق خاصیت کانولوشن در DTFT:
 $$ y[n] = x[n] * h[n] → Y(e^{jω}) = X(e^{jω}) H(e^{jω}) $$


## سؤال ۴: تبدیل فوریه سیگنال محدود

اگر سیگنال زیر داده شده باشد:
 $$ x[n] = u[n] - u[n-5] $$
تبدیل فوریه گسسته در زمان آن را بیابید.

پاسخ:
با استفاده از تبدیل فوریه دنباله پله واحد، داریم:
 $$ X(e^{jω}) = Σ e^{-jωn} FOR n = 0.4 $$
که حاصل سری هندسی است:
$$ X(e^{jω}) = (1 - e^{-j5ω}) / (1 - e^{-jω}) $$


## سؤال ۵: تبدیل فوریه سیگنال نمایی

تبدیل فوریه گسسته در زمان سیگنال$$  x[n] = α^n u[n] $$ را محاسبه کنید، به شرطی که |α| < 1 باشد.

پاسخ:
با استفاده از تعریف DTFT:
$$ X(e^{jω}) = Σ α^n e^{-jωn}  FOR n ≥ 0  $$
این یک سری هندسی است که مقدار آن برابر است با:
$$ X(e^{jω}) = 1 / (1 - α e^{-jω})، FOR  |α| < 1 $$


## سؤال ۶: محاسبه DTFT از یک تابع سینوسی

سیگنال  $$ x[n] = cos(ω0 n) $$ داده شده است. تبدیل فوریه گسسته در زمان آن را محاسبه کنید

پاسخ:
با استفاده از فرمول اویلر:
$$ cos(ω0 n) = (e^{jω0 n} + e^{-jω0 n}) / 2 $$
پس تبدیل فوریه گسسته آن برابر است با:
$$ X(e^{jω}) = π [ δ(ω - ω0) + δ(ω + ω0) ] $$


## سؤال ۷: خاصیت دوره‌ای بودن DTFT

ثابت کنید که تبدیل فوریه گسسته در زمان یک سیگنال گسسته، یک تابع دوره‌ای با دوره 2π است.

پاسخ:
با توجه به تعریف DTFT:
$$ X(e^{jω}) = Σ x[n] e^{-jωn} $$
اگر ω را با ω + 2π جایگزین کنیم، مشاهده می‌شود که معادله تغییر نمی‌کند، بنابراین:
$$ X(e^{j(ω+2π)}) = X(e^{jω}) $$
پس DTFT یک تابع دوره‌ای با دوره 2π است.


## سؤال ۸: محاسبه DTFT برای سیگنال مستطیلی

سیگنال x[n] به صورت یک پنجره مستطیلی با طول N داده شده است:
x[n] = 1 برای 0 ≤ n < N و صفر در غیر این صورت.
تبدیل فوریه گسسته در زمان آن را به دست آورید.

پاسخ:
$$ X(e^{jω}) = Σ_{n=0}^{N-1} e^{-jωn} $$
این یک سری هندسی است که مقدار آن برابر است با:
$$ X(e^{jω}) = (1 - e^{-jNω}) / (1 - e^{-jω}) $$
