 ## 1-1-2 نمایش سیگنالهای گسسته در زمان بر حسب ضربه  ##
 ایده اصلی نمایش سیگنالهای گسته در زمان بر حسب ضربه واحد گسته در زمان ، تصور سیگنال به  صورت مجموع یک رشته ضربه جابجا شده است. 
 برای تبدیل این تصویر شهودی به نمایش ریاضی  سیگنال x[n] شکل ۲-۱ (الف) را در نظر بگیرید. در بخشهای دیگر این شکل پنج ضربه جابجا شده ، با  ضریبهای مختلف نشان داده شده است، ضریب هر ضربه مقدار x[n]  در لحظه رخ دادن آن ضربه است. 
 $$ x[-1]δ[n+1] =x[-1],n=-1,0 , n≠-1 $$ 
$$ x[0]δ[n] =x[0],n=0,0  , n≠0 $$
$$ x[1]δ[n-1] =x[1],n=1,0  , n≠1 $$

بنابراین در فاصله بین منفی 2 تا 2 با مجموع پنج دنباله نشان داده شده در شکل برابر است. در حالت کلی  با در نظر گرفتن ضربه های جابجا شده و تغییر مقیاس یافته ، می توان نوشت  

$$ [n]=⋯+x[-3]δ[n+3]+x[-2]δ[n+2]+x[-1]δ[n+1]+x[0]δ[n]+x[1]δ[n-1]+x[2]δ[n-2]+x[3]δ[n-3]+⋯ $$

به ازای هر ،تنها یکی از جملات طرف راست معادله (2-1) مخالف صفرست ، و ضریب آن جمله دقیقاً برابراست. معادله فوق را می توان به صورت فشرده تر زیر نوشت  
$$ x[n]=\sum_{k=-\infty}^{\infty} x[k]\delta[n-k] $$

معادله فوق نمایش یک دنباله دلخواه ترکیب خطی ضربه های واحد جابه جاشده در زمان یعنی را بیان میکند ؛ ضرایب این ترکیب خطی ها هستند. برای مثال  یعنی پله واحد را در نظر بگیرید. در این حالت چون به ازای معادله (۲-۲) به صورت زیر در می آید

یعنی  خروجی سیستم به ورودی  است. پس معادله (۲-۳) برای سیستمهای خطی تغییر ناپذیر با زمان به  صورت زیر در می آید  (۶-۲)  

$$ y[n]=\sum_{k=-\infty}^{\infty} x[k] h[n-k] $$

نتيجة فوق جمع کانولوشن نامیده می شود. عمل طرف راست معادله (۲-۶) کانولوشن دنباله های  در  نام دارد و آن را با نماد زیر نشان می دهیم  

$$ y[n]=x[n]*h[n] $$

توجه کنید که سیستم خطی تغییر ناپذیر با زمان به یک ورودی دلخواه را بر حسب پاسخ آن سیستم به ضربة  واحد بیان می کند . پس می بینیم که یک سیستم تغییر ناپذیر با زمان با پاسخش به یک سیگنال تنها ، یعنی پاسخش به ضربه  واحد ، به طور کامل مشخص میشود.  
تعبیر معادله (۲-۶) با تعبیری که برای معادله (2-3) بیان کردیم مشابه است ، ولی در مورد سیستم خطی تغییر ناپذیر با زمان  پاسخ ناشی از ورودی اعمال شده در زمان برابر است ، که گونه تغییر مقیاس یافته و  جابجا شده  میباشد. خروجی واقعی جمع تمام این گونه پاسخهاست .
